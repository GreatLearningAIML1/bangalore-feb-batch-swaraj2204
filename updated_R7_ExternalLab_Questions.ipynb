### Build a DNN using Keras with `RELU` and `ADAM`

#### Load tensorflow

```python
import tensorflow as tf
```

#### Collect Fashion mnist data from tf.keras.datasets 

```python
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

#### Change train and test labels into one-hot vectors

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
y_train.shape
```

#### Build the Graph

#### Initialize model, reshape & normalize data

```python
#Normalize Data
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Dropout, Reshape
```

```python
model = Sequential()
model.add(Reshape((784,),input_shape=(28,28,)))
```

#### Add two fully connected layers with 200 and 100 neurons respectively with `relu` activations. Add a dropout layer with `p=0.25`

```python
model.add(Dense(200, activation='relu', name='Layer1'))
model.add(Dense(100, activation='relu', name='Layer2'))
model.add(Dropout(0.25))
```

### Add the output layer with a fully connected layer with 10 neurons with `softmax` activation. Use `categorical_crossentropy` loss and `adam` optimizer and train the network. And, report the final validation.

```python
model.add(Dense(10, activation='softmax', name='Output'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, validation_split=0.25 , epochs=50, batch_size=1000, verbose=0)
```

```python
train_score = model.evaluate(x_train, y_train, verbose=0)
score = model.evaluate(x_test, y_test, verbose=0)
print(f'Train Score :- {train_score[1]*100} & Test Score :- {score[1]*100}')
```
